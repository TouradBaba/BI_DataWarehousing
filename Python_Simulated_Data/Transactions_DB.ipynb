```python
import mysql.connector
from faker import Faker
import random
import datetime

# Connect to the MySQL database
connection = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="**********",
    database="transactions_db"
)

# Create a cursor object to execute SQL queries
cursor = connection.cursor()


# Create a cursor object to execute SQL queries
cursor = connection.cursor()

# Initialize Faker to generate realistic data
fake = Faker()

# Function to generate random dates within a given range
def random_date(start_date, end_date):
    return fake.date_time_between(start_date=start_date, end_date=end_date)

# Function to generate fictive data for products
def generate_products(num_records):
    products = []
    for _ in range(num_records):
        product = (
            fake.word(ext_word_list=['laptop', 'smartphone', 'tablet', 'monitor', 'printer']),
            fake.random_element(elements=('Electronics', 'Computers', 'Accessories')),
            round(random.uniform(50, 2000), 2)  # Price
        )
        products.append(product)
    return products

# Function to generate fictive data for customers
def generate_customers(num_records):
    customers = []
    for _ in range(num_records):
        customer = (
            fake.first_name(),
            fake.last_name(),
            fake.email(),
            fake.phone_number(),
            fake.address()
        )
        customers.append(customer)
    return customers


# Generate fictive data for products and customers
products_data = generate_products(50)
customers_data = generate_customers(100)

# SQL statements to insert data into tables
insert_products_query = "INSERT INTO Products (ProductName, Category, Price) VALUES (%s, %s, %s)"
insert_customers_query = "INSERT INTO Customers (FirstName, LastName, Email, Phone, Address) VALUES (%s, %s, %s, %s, %s)"

# Insert data into Products table
cursor.executemany(insert_products_query, products_data)

# Insert data into Customers table
cursor.executemany(insert_customers_query, customers_data)


# Function to generate fictive data for sales transactions
def generate_transactions(num_records, customers, products):
    transactions = []
    for _ in range(num_records):
        customer_id = random.choice(customers)[0]  # Accessing the first element (CustomerID) in the tuple
        product_id = random.choice(products)[0]   # Accessing the first element (ProductID) in the tuple
        # Generate a random order date within the past year
        order_date = fake.date_time_between(start_date='-1y', end_date='now').date()
        transaction1 = (
            customer_id,  # CustomerID
            product_id,   # ProductID
            order_date,   # OrderDate
            random.randint(1, 10),  # Quantity
            random.uniform(50, 2000)  # UnitPrice (adjust the range as needed)
        )
        transactions.append(transaction1)
    return transactions


# Retrieve existing data from Products and Customers tables
cursor.execute("SELECT ProductID FROM Products")
products = cursor.fetchall()
cursor.execute("SELECT CustomerID FROM Customers")
customers = cursor.fetchall()

# SQL statement to insert data into SalesTransactions table
insert_transactions_query = "INSERT INTO Transactions (CustomerID, ProductID, OrderDate, Quantity, UnitPrice) VALUES (%s, %s, %s, %s, %s)"

# Generate fictive data for sales transactions
transactions_data = generate_transactions(500, customers, products)


# Insert data into Transactions table
cursor.executemany(insert_transactions_query, transactions_data)

# Commit the changes
connection.commit()

# Close the cursor and connection
cursor.close()
connection.close()

print("Data populated successfully!")
```
